```python
import numpy as np
import pandas as pd
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')   
```

```python
dengue = pd.read_csv("E:\Portable_data\Codes\dengue analysis\dengue.csv")
dengue
```

<h1>1. Comparison of Cases and Deaths</h1>


```python

dengue[['Dengue_Cases','Dengue_Deaths']].sum().plot(kind='bar', figsize=(6,4))
plt.title("Comparison: Total Cases vs Deaths")
plt.ylabel("Total Count")
plt.grid(alpha=0.3)
plt.show()
```

<h1>Insights</h1>
<h3>Insight 1: Although dengue cases are overwhelmingly higher than deaths the proportional gap suggests effective treatment and case management across the years.
 </h3>
<h3>Insight 2: The consistent low fatality despite high infections reflects improvements in early detection and community awareness. </h3>

<h1>2. Highest Dengue Month per Year</h1>

```python

max_months = dengue.loc[dengue.groupby('Year')['Dengue_Cases'].idxmax()][['Year','Month','Dengue_Cases']]
print(max_months)
```

<h1>Insights</h1>
<h3>Insight 1: The table consistently shows August and September as peak months for most years reinforcing the strong link between rainfall and dengue spread.
</h3>
<h3>Insight 2: Identifying these peak periods helps public health agencies target preventive spraying and awareness drives more effectively. </h3>

<h1>3. Correlation Between Cases and Deaths</h1>

```python

correlation = dengue['Dengue_Cases'].corr(dengue['Dengue_Deaths'])
print(f"Correlation between Dengue Cases and Deaths: {correlation:.2f}")
```

<h1>Insights</h1>
<h3>Insight 1: The correlation value close to +1 indicates a strong positive relationship meaning that as dengue cases increase deaths tend to rise as well.
</h3>
<h3>Insight 2: This emphasizes that managing outbreaks early can directly reduce fatalities showing the importance of fast response systems.</h3>

<h1>4. Regions with Consistently High Cases</h1>

```python

dengue.groupby('Region')['Dengue_Cases'].sum().sort_values(ascending=False).head(10).plot(kind='bar', figsize=(10,5))
plt.title("Regions with Highest Total Dengue Cases")
plt.ylabel("Number of Cases")
plt.xticks(rotation=45)
plt.grid(alpha=0.3)
plt.show()
```

<h1>Insights</h1>
<h3>Insight 1:NCR, Region IV-A, and Region III consistently top the list for dengue cases likely due to population density and environmental conditions favoring mosquitoes.
 </h3>
<h3>Insight 2: These findings highlight that metropolitan areas remain major risk zones and urban sanitation improvements could significantly lower cases. </h3>

<h1>5. Case Fatality Rate per Region</h1>

```python

fatality_rate = dengue.groupby('Region').apply(lambda x: x['Dengue_Deaths'].sum() / x['Dengue_Cases'].sum() * 100)
fatality_rate.sort_values(ascending=False).plot(kind='bar', color='orange', figsize=(10,5))
plt.title("Case Fatality Rate (%) per Region")
plt.ylabel("Fatality Rate (%)")
plt.xticks(rotation=45)
plt.grid(alpha=0.3)
plt.show()
```

<h1>Insights</h1>
<h3>Insight 1: Some smaller regions show surprisingly high fatality rates suggesting possible delays in diagnosis or limited access to healthcare facilities.
 </h3>
<h3>Insight 2: Meanwhile, larger regions with more hospitals maintain lower fatality percentages reflecting the benefits of healthcare infrastructure and readiness. </h3>
